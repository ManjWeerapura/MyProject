```python
import pandas as pd
import numpy as np
```

```python
trading_portfolio = pd.ExcelFile('trading_portfolio.xlsx')
trading_portfolio #load the Excel file with the data
```

```python
trading_portfolio.sheet_names # output of this is a list of all the sheets within the Excel file
# catherefore, you can loop through the sheet names and load them into dataframes individually
```

```python
# read one of the excel file sheets into a dataframe
df = pd.read_excel('trading_portfolio.xlsx', sheet_name = trading_portfolio.sheet_names[1])
```

```python
df
```

```python
df.info()
```

```python
df = df[:-1]
```

```python
df
```

```python
df[['Price','Open', 'High', 'Low', 'Change %']] = df[['Price','Open', 'High', 'Low', 'Change %']].apply(pd.to_numeric)
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df
```

```python
df['Mean'] = df['Price'].rolling(20).mean()
```

```python
df
```

```python
df['Mean+2STD'] = df['Mean']+df['Price'].rolling(20).std()*2
```

```python
df
```

```python
df['Mean-2STD'] = df['Mean']-df['Price'].rolling(20).std()*2
```

```python
df
```

```python
# drop the rwos will null values
df = df.dropna(axis=0)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.Mean.dtype
```

```python
plt.figure(figsize = (15,7))
sns.lineplot(x = df['Date'], y = df['Mean'], color = 'orange', linewidth = 1)
sns.lineplot(x = df['Date'], y = df['Mean+2STD'], color = 'blue', linewidth = 2)
sns.lineplot(x = df['Date'], y = df['Mean-2STD'], color = 'red', linewidth = 2)
sns.lineplot(x = df['Date'], y = df['Price'], color = 'green', linewidth = 1)
plt.fill_between(x=df['Date'], y1 = df['Mean+2STD'], y2 = df['Mean-2STD'], color= 'grey', alpha = 0.3)
plt.legend(['Mean', 'Mean + 2STD', 'Mean - 2STD', 'Closing Price'])
```
